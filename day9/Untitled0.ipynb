```python
"""Write a Python Program to get 5 integer numbers from user and store in a list, display that
number using for loop, every number should display with its square."""


lst=[]
for i in range(5):
  lst.append(int(input("Enter a number: ")))
for i in range(5):
  print(lst[i])

print("The Square Value:")
for i in range(5):
  x=lst[i]
  lst1= x**2
  print(lst1)



```

```python
"""Write a Python Program to get 5 any words from user, display that word which ends with ‘n’
character."""
lst=[]
for i in range(5):

  lst.append(input("Enter a name: "))

for i in range(5):
  print(str(lst[i]).endswith('n'))

```

```python
"""Write a Python Program to Get Two Set from User to Find
Intersection of those Set."""
str1 = set(input("Please enter a string: "))
str2 = set(input("Please enter a string: "))
x= str1.intersection(str2)
print("The intersection is: ",x)
```

```python
"""Write a Python Program to Get Two Set from User and Check which
Set is Subset of Other?"""
str1 = set(input("Please enter a string: "))
str2 = set(input("Please enter a string: "))
x=str1.issubset(str2)
print(x)
```
