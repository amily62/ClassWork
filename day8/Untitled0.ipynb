```python
"""Write a Python Program to get 5 Students Name from user and Store in List, Display all the
Students Name with Uppercase Letters."""

list=[]
for i in range(5):
  list.append(input(f"Enter name {i}: "))

for i in range(5):
  list[i]=(list[i].upper())

print("The List = ",list)

```
