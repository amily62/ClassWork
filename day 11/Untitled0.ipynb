```python
class Student:
    _id = 389
    _name = "Mahmudul Hasan"

    def _show(self):
        print("ID: ", self._id)

class Jafri(Student):
    def showDetails(self):
        print("Name:", self._name)
        self._show()

obj = Jafri()
obj.showDetails()
```
