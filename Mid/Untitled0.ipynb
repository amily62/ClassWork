```python
num = 15
flag = 0
for i in range(2,num):
  if num%i==0:
    flag = 1
    break
if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

```python
import random
random_number = random.randint(1, 100)

attempts = 0

while attempts < 5:
    guess = int(input("Guess the number (between 1 and 100): "))
    attempts += 1

    if guess == random_number:
        print("Congratulations! You guessed the number.")
        break
    elif guess < random_number:
        print("Too low!")
    else:
        print("Too high!")
else:
    print(f"Sorry, the number was {random_number}.")
```

```python
from statistics import mean, median, mode

# Input the list of numbers
numbers = list(map(float, input("Enter space-separated numbers: ").split()))

# Calculate and print the results
print("Mean:", mean(numbers))
print("Median:", median(numbers))

# Handle mode in case of no unique mode
try:
    print("Mode:", mode(numbers))
except:
    print("Mode: No unique mode")
```

```python
list1 = list(map(int, input("Enter elements of the first list: ").split()))
list2 = list(map(int, input("Enter elements of the second list: ").split()))

set1, set2 = set(list1), set(list2)

print("Union:", list(set1 | set2))
print("Intersection:", list(set1 & set2))
print("Symmetric Difference:", list(set1 ^ set2))
```

```python
input_file = "input.txt"
output_file = "output.txt"

with open(input_file, "r") as file:
    text = file.read()

word_count = {}
for word in text.split():
    word = word.lower().strip(",.!?")
    word_count[word] = word_count.get(word, 0) + 1

with open(output_file, "w") as file:
    for word, count in sorted(word_count.items()):
        file.write(f"{word}: {count}\n")

print(f"Word frequencies have been saved to {output_file}.")
```
